```python
import os
import numpy
```

```python
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname = distance_file, delimiter = ',', dtype = 'unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
print(data)
```

```python
print(headers)
```

```python
import matplotlib.pyplot as plt
```

```python
sample = headers[1]
sample2 = headers[2]

plt.plot(data[:,1], label = sample, color = 'm')
plt.plot(data[:,2], label = sample2, color = 'g')
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.title('Pregante no bueno')
plt.legend()
plt.savefig(F'two_samples.png', dpi = 300)
```

```python
num_columns = len(headers)
print(headers)
print(F'The number of columbs is {num_columns}')
```

```python
num_columns = len(headers)
print(headers)
print(F'The number of columns is {num_columns}')
color = [0, 'r', 'g', 'm', 'c']

for col in range(1, num_columns):
    plt.figure()
    plt.plot(data[:,col], label = headers[col], color = color[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.title('Pregante no bueno')
    plt.legend()
    plt.savefig(F'{sample}{col}', dpi = 200)
```
